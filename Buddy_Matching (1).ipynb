```python
#Q3,Q6
#Q25,17
#23,Q20
#Q27,21
import pandas as pd
from difflib import SequenceMatcher

# Load the data
buddy_df = pd.read_csv('Buddy.csv')
new_hire_df = pd.read_csv('New_Hire.csv')

def find_best_buddy(new_hire):
    best_match = None
    best_ratio = 0

    for _, buddy in buddy_df.iterrows():
        ratio = (
            SequenceMatcher(None, str(new_hire['Q6']), str(buddy['Q3'])).ratio() +
            SequenceMatcher(None, str(new_hire['Q17']), str(buddy['Q25'])).ratio() +
            SequenceMatcher(None, str(new_hire['Q20']), str(buddy['Q23'])).ratio() +
            SequenceMatcher(None, str(new_hire['Q21']), str(buddy['Q27'])).ratio()
        )

        if ratio > best_ratio:
            best_match = buddy
            best_ratio = ratio

    return best_match, best_ratio

for _, new_hire in new_hire_df.iterrows():
    best_buddy, best_ratio = find_best_buddy(new_hire)
    print(f"New hire: {new_hire['Q1']} | Best buddy: {best_buddy['Q1']} (Match ratio: {best_ratio:.2f})")
```

```python
import pandas as pd
from rapidfuzz.fuzz import ratio
from rapidfuzz.process import extractOne
from fuzzywuzzy import fuzz
# Load the data
buddy_df = pd.read_csv('Buddy.csv')
new_hire_df = pd.read_csv('New_Hire.csv')

def find_best_buddy_fuzzy(new_hire):
    best_match = None
    best_ratio = 0

    for _, buddy in buddy_df.iterrows():
        match_ratio = (
            ratio(str(new_hire['Q6']), str(buddy['Q3'])) +
            ratio(str(new_hire['Q17']), str(buddy['Q25'])) +
            ratio(str(new_hire['Q20']), str(buddy['Q23'])) +
            ratio(str(new_hire['Q21']), str(buddy['Q27']))
        )

        if match_ratio > best_ratio:
            best_match = buddy
            best_ratio = match_ratio

    return best_match, best_ratio

for _, new_hire in new_hire_df.iterrows():
    best_buddy, best_ratio = find_best_buddy_fuzzy(new_hire)
    print(f"New hire: {new_hire['Q1']} | Best buddy: {best_buddy['Q1']} (Match ratio: {best_ratio:.2f})")
```
